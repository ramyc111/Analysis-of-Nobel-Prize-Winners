# High Proportion of Missing Values:
 The overallMotivation column has 956 missing values out of a total of 969 entries. This means that more than 98% of the entries are missing data. The correlation heatmap indicates that overallMotivation has little to no correlation with other columns. This means its missingness is likely independent, making imputation with internal dataset information difficult and unreliable.
Limited Analysis Impact:
 The specific motivations provided in this column do not significantly impact general demographic or prize-related trends. Since similar details can often be derived from the motivation column (which is more populated), dropping overallMotivation will not lead to substantial information loss.
Given these reasons, it is logical to drop the overallMotivation column to simplify the dataset and focus on meaningful, analyzable data.

Extract and preprocess biographical descriptions to prepare for NLP tasks.
Apply tokenization, stopword removal, and lemmatization using SpaCy or NLTK to ensure text is ready for entity recognition.


```python
import pandas as pd
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Text Preprocessing for NLP using SpaCy
nlp = spacy.load('en_core_web_sm')

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return ' '.join(tokens)

nobel_data_cleaned['processed_motivation'] = nobel_data_cleaned['motivation'].fillna('').apply(preprocess_text)
```

# What are the age trends across Nobel laureates?

```python
# Age Trends: Calculate the age at award receipt, analyze category-wise distributions, and observe changes in the age profile over decades (for individuals only)
individuals_data = nobel_data_cleaned[nobel_data_cleaned['gender'] != 'org']
individuals_data['award_year'] = pd.to_numeric(individuals_data['year'], errors='coerce')
individuals_data['age_at_award'] = individuals_data['award_year'] - individuals_data['born'].dt.year

# Drop rows with invalid ages
individuals_data = individuals_data[individuals_data['age_at_award'] >= 0]

# Plot category-wise distribution of age at award
plt.figure(figsize=(12, 8))
individuals_data.boxplot(column='age_at_award', by='category', grid=False)
plt.title('Age at Award by Category')
plt.suptitle('')
plt.xlabel('Category')
plt.ylabel('Age at Award')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analyze changes in the age profile over decades
individuals_data['decade'] = (individuals_data['award_year'] // 10) * 10
avg_age_per_decade = individuals_data.groupby('decade')['age_at_award'].mean()

# Plot average age at award over decades
plt.figure(figsize=(10, 6))
avg_age_per_decade.plot(kind='line', marker='o', color='coral')
plt.title('Average Age at Award Over Decades')
plt.xlabel('Decade')
plt.ylabel('Average Age at Award')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Number of Nobel Prizes Awarded by Gender (1996-2016)

```python
# Filter data for the last 20 years
latest_year = nobel_data_cleaned['award_year'].max()
last_20_years_data = nobel_data_cleaned[(nobel_data_cleaned['award_year'] >= 1996) & (nobel_data_cleaned['award_year'] <= latest_year)]
gender_awards = last_20_years_data.groupby(['award_year', 'gender']).size().unstack().fillna(0)

# Plot the number of Nobel Prizes awarded by gender (1996-2016)
plt.figure(figsize=(12, 6))
gender_awards.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Number of Nobel Prizes Awarded by Gender (2004-2024 or Latest Available Year)')
plt.xlabel('Year')
plt.ylabel('Number of Prizes Awarded')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
